```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,5,13,33,46,25]).reshape((-1,1))
y=np.array([3,6,9,12,14,18])
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
SLR=LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Red',marker='*')

print("MSE:",mse)
print("intercept:", SLR.intercept_)
print("slope:",SLR.coef_)
```

**generating random numbers in linear regression**

```python
import random
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


random.seed(42)
num_points = 50
x = [random.uniform(0, 10) for _ in range(num_points)]
y = [2 * xi + random.gauss(0, 1) for xi in x]


x = np.array(x).reshape(-1, 1)
y = np.array(y)


model = LinearRegression()


model.fit(x,y)


y_pred = model.predict(x)


plt.scatter(x, y, label='Original Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()


print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Read the Excel file
excel_file = "/content/drive/MyDrive/values.xlsx"
data = pd.read_excel(excel_file)

# Assuming your Excel columns are named 'x' and 'y'
x = data['x'].values.reshape(-1, 1)
y = data['y'].values

# Create a linear regression model
model = LinearRegression()
model.fit(x, y)

# Predict y values using the model
y_pred = model.predict(x)

# Plot the original data and the regression line
plt.scatter(x, y, label='Original Data')
plt.plot(x, y_pred, color='red', label='Linear Regression')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression')
plt.legend()
plt.show()

# Print the slope (coefficients) and intercept of the regression line
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import numpy as np

# Generate some random data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Initialize parameters
theta = np.random.randn(2, 1)

# Define hyperparameters
learning_rate = 0.1
num_iterations = 1000

# Perform gradient descent
for iteration in range(num_iterations):
    gradients = -2 * X.T.dot(y - X.dot(theta)) / len(X)
    theta -= learning_rate * gradients

# Print the final parameters
print("Optimized theta:", theta)
```
