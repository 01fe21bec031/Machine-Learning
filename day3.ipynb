```python
def add(*bunch):
   sum=0
   for i in bunch:
     sum+= i
   return sum
print(add(1,2,3,4,5))
```

```python
def greet(**kwargs):
  print("Hello")
  for key,value in kwargs.items():
    print(value, end=" ")

greet(name="Supriya", usn="SG9181", branch="ECE")
```
